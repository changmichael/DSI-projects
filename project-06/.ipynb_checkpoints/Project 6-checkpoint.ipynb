```python
def get_top_250():
    # Getting the list of the top 250 movies to start off with
    response = requests.get('http://www.imdb.com/chart/top')
    html = response.text
    entries = re.findall("<a href.*?/title/(.*?)/", html)
    return list(set(entries))

def get_entry(entry):
    # Getting the individual entry for each movie
    res = requests.get('http://www.omdbapi.com/?i='+entry)
    if res.status_code != 200:
        print(entry, res.status_code)
    else:
        print('.', end=" " )
    try:
        j = json.loads(res.text)
    except ValueError:
        j = None
    return j

def intminutes(x):
    # Changing the format of the Runtime column, also intergerizing it
    y = x.replace('min', '').strip()
    return int(y)

def intvotes(x):
    # Changing the format of the votes columns, also integerizing it
    y = x.replace(',', '').strip()
    return int(y)

def ratingCleaner(x):
    # Cleaning up the values in the Rated column so that they're more concise
    y = str(x).split()[0]
    if y == 'Rated':
        return str(x).split()[1]
    elif y == 'NOT':
        return x
    else:
        return y

def ratingCombiner(x):
    # Combining different ratings with each other to reduce the number
    if x == 'M':
        return 'X'
    elif x == 'UNRATED':
        return 'NOT RATED'
    elif x == 'PASSED':
        return 'NOT RATED'
    elif x == 'N/A':
        return 'NOT RATED'
    elif x == 'APPROVED':
        return 'G'
    else:
        return x
    
def get_soup_from_url(url):
    # Get Beautiful Soup for a given webpage
    response = requests.get(url)
    page = response.text
    soup = BeautifulSoup(page,'lxml')
    return soup

def get_movie_info(url):
    #  Scraping the movie title (needed for merging data later), number of user ratings and 
    #  number of critic ratings
    soup = get_soup_from_url(url)
    results = []
    title_string = soup.find('h1').text
    title = str(title_string.split('(')[0].strip().encode('utf8'))
    userRatings = str(soup.find('span',{'itemprop':'reviewCount'}).text.split()[0])
    criticRatings = int(soup.find('span',{'itemprop':'reviewCount'}).findNextSibling().text.split()[0])
    results.append(title)
    results.append(userRatings)
    results.append(criticRatings)
    return results  
```

```python
# Getting the list of top 250 movies and the list of their links
links=[]
dfimdb = pd.DataFrame(imdb.top_250())
top_250_url = 'http://www.imdb.com/chart/top?ref_=nv_mv_250_6'
soup_250= get_soup_from_url(top_250_url)

for x in dfimdb['tconst']:
    url1 = 'http://www.imdb.com/title/'
    url2 = '/?ref_=nv_sr_1'
    link = url1 + x + url2
    links.append(str(link))
```

```python
# The actual scrape - scraping the info, putting it into a table called (appropriately) 'table'
table = []
for url in links:
    info = get_movie_info(url)
    table.append(info)
```

```python
# Making the dataframe for the scraped data.  We'll be concat-ing it later to the main df.  
scrapeddf = pd.DataFrame(table, columns = ['Title','userRatings','criticRatings'])
```

```python
# Checking to make sure we got all 250 movies
entries = get_top_250()
len(entries)
```

```python
# Using the API to get all the other info
entries_dict_list = [get_entry(e) for e in entries]
```

```python
# making the main dataframe with all the info from the API, all info is input.  
df = pd.DataFrame(entries_dict_list)
```

```python
# Merging web-scraped stuff and api-ed stuff
df = pd.merge(df, scrapeddf)
```

```python
# Changing data into nan's as appropriate, then taking a count of how many in each column
df = df.replace('N/A', np.nan)
df.isnull().sum()
```

```python
# Lots of cleaning and munging. Filling in some values, changing formats, integerizing, and 
# also dropping a lot of columns which either aren't relevant or have too many missing values.
df.Language.fillna('english', inplace=True)
df.Rated.fillna('NOT RATED', inplace=True)
df.Released = pd.to_datetime(df.Released)
df.Runtime = df.Runtime.apply(intminutes)
df.Year = df.Year.astype(int)
df.imdbRating = df.imdbRating.astype(float)
df.imdbVotes = df.imdbVotes.apply(intvotes)
df.userRatings = df.userRatings.apply(intvotes)
df.userRatings = df.userRatings.apply(lambda x: int(x))
df.criticRatings = df.criticRatings.apply(lambda x: int(x))
df['Rated'] = df['Rated'].apply(lambda x: ratingCleaner(x))
df['Rated'] = df['Rated'].apply(lambda x: ratingCombiner(x))
df.drop('Awards',1,inplace=True)
df.drop('Metascore',1,inplace=True)
df.drop('Poster',1,inplace=True)
df.drop('Released',1,inplace=True)
df.drop('Response',1,inplace=True)
df.drop('Type',1,inplace=True)
df.drop('Writer',1,inplace=True)
df.drop('imdbID',1,inplace=True)
```

# Regressors

```python
# Setting things up to do regressors all at once, also setting up the difference regressor 
# tree and boost models, and plotting functions.  

cv = KFold(len(y), shuffle=True)
dt = DecisionTreeRegressor(max_depth=5)
bdt = BaggingRegressor(DecisionTreeRegressor())
rf = RandomForestRegressor()
et = ExtraTreesRegressor()
ab = AdaBoostRegressor()
gb = GradientBoostingRegressor()

def score(model, name):
    s = cross_val_score(model, X, y, cv=cv, n_jobs=-1)
    print("{} Score:\t{:0.3} ± {:0.3}".format(name, s.mean().round(3), s.std().round(3)))

def scoreNewX(model, name):
    s = cross_val_score(model, newX, y, cv=cv, n_jobs=-1)
    print("{} Score:\t{:0.3} ± {:0.3}".format(name, s.mean().round(3), s.std().round(3)))

def scoreNewerX(model, name):
    s = cross_val_score(model, newerX, y, cv=cv, n_jobs=-1)
    print("{} Score:\t{:0.3} ± {:0.3}".format(name, s.mean().round(3), s.std().round(3)))
        
def do_plot(model, m=None):
    for fold, color in zip(cv, ['r', 'g', 'b']):
        X_train = X.iloc[fold[0]]
        X_test  = X.iloc[fold[1]]
        y_train = y.iloc[fold[0]]
        y_test  = y.iloc[fold[1]]
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        plt.scatter(y_test, y_pred, color=color)
        plt.plot([8.5,9.5],[8.5,9.5])
        plt.text(9,9.5, "R2: "+str(m), fontsize=20, )
```

```python
# Getting the scores for the original X
score(dt, "Decision Tree")
score(bdt, "BaggingRegressor")
score(rf, "Random Forest")
score(et, "Extra Trees")
score(ab, "Adaboost ")
score(gb, "GradientBoosting")
```

# Grid search fo Decision Tree Regressor
We take a look at grid search for the decision tree regressor; we also tried it for bagging, but the calculation time appeared onerous (waiting 10+ minutes with no results).  We just post the results from the Decision Tree Regressor grid search as it's quick, straightforward and easy to interpret.

```python
# I set these parameters but i'm sure there's probably a more optimial set of parameters to 
# choose.  
PARAMETERS = {'max_depth':[3,5,10,20], 'max_features':[1,2,3,'auto'], 
              'max_leaf_nodes':[5,6,7,8,9,10], 'min_samples_leaf':[1,3,5,7,10],
              'min_samples_split':[2,5,7]}
SCORING = 'mean_squared_error'
```

```python
# Grid Searching.  With decision tree regressor.  Nice and quick.  
model = DecisionTreeRegressor()
clf = GridSearchCV(model, PARAMETERS, scoring=SCORING, verbose=True, n_jobs=-1)
clf.fit(newX, y)
```

```python
# Just printing the results in a nice easy format to digest. 
print("Best model: \n \n {}".format(clf.best_estimator_))
print 
print("Best score was {}".format(clf.best_score_))
print(np.sqrt(-clf.best_score_))
```

```python
# just printing out the best parameters as spat out by grid search
clf.best_params_
```

```python
# and printing out the best score as spat out by grid search
clf.best_score_
```

```python
# Shamelessly using this code from class - to visualize the grid search itself. We see the 
# best score at depth of 5 and is best visualized using the below graphing functions.

all_scores = []
best_score = -1
best_depth = 0
for i in range(1, 9):
    treereg = DecisionTreeRegressor(max_depth=i, random_state=1)
    scores = cross_val_score(treereg, newX, y, cv=3, scoring='mean_squared_error')
    current_score = np.mean(np.sqrt(-scores))
    # If the score mean is better than the current best, or best is the default (-1), then update!
    if current_score < best_score or best_score == -1:
        best_score = current_score
        best_depth = i
    # store to plot anyway!
    all_scores.append(current_score)
    
print("Best score: %s" % best_score)
print("Best depth: %s" % best_depth)

# now actually fit the model
treereg = DecisionTreeRegressor(max_depth=best_depth, random_state=1)
treereg.fit(newX, y)

plt.figure()
plt.plot(range(1, 9), all_scores)
plt.xlabel('x=max tree depth')
plt.ylabel('training error')
```

So we see that one of the parameters that we can set provides the best results, but we see that depthes of 3 and 4 also work similarly well.  In any case, we had set it at 5 in the beginning as a lark and just got really really lucky.

```python
all_scores = []
best_score = -1
best_depth = 0
for i in range(1, 9):
    treereg = DecisionTreeRegressor(max_depth=i, random_state=1)
    scores = cross_val_score(treereg, newerX, y, cv=3, scoring='mean_squared_error')
    current_score = np.mean(np.sqrt(-scores))
    # If the score mean is better than the current best, or best is the default (-1), then update!
    if current_score < best_score or best_score == -1:
        best_score = current_score
        best_depth = i
    # store to plot anyway!
    all_scores.append(current_score)
    
print("Best score: %s" % best_score)
print("Best depth: %s" % best_depth)

# now actually fit the model
treereg = DecisionTreeRegressor(max_depth=best_depth, random_state=1)
treereg.fit(newerX, y)

plt.figure()
plt.plot(range(1, 9), all_scores)
plt.xlabel('x=max tree depth')
plt.ylabel('training error')
```
